```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
from scipy.stats import norm
from statsmodels.stats.weightstats import ztest
```

```python
df = pd.read_csv('titanic_train.csv')
```

```python
# sobrevivientes
x1 = df[df['Survived']==1]['Fare'].dropna().to_numpy()
# No taaaan sobrevivientes
x2 = df[df['Survived']==0]['Fare'].dropna().to_numpy()
```

```python
sns.kdeplot(x1, label='Survived')
sns.kdeplot(x2, label='Did not Survived')
plt.legend();
```

```python
x1.mean(), x2.mean()
```

```python
ztest(x1,x2)
```

Con $10^{-15}$ de p-value podemos decir con facilidad que las medias de ambos grupos son **significativamente** distintas.   
Rechazamos la hipótesis nula $x1=x2$.  
Tenemos evidencia *estadística* de que el costo del voleto sí afectó la probabilidad de supervivencia.
